```python
series = (
    googl
        .reset_index()[['date', '2. high']]
        .rename({"2. high": "high"}, axis=1)
)
```

```python
def add_window(series, window=10):
    for period in range(1, window + 1):
        series.insert(len(series.columns), f"t_{period}", series.high.shift(period))
    return series
```
